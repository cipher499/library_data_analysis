```python
# cleaned in excel

df = pd.read_excel('gramsci2_c1.xlsx')
df.head()
```

```python
df.isnull().sum()
```

```python
df['PURPOSE'].value_counts()
```

```python
# Bank appears twice in the list of unique categories
df.replace('Bank ', 'Bank', inplace=True)
```

```python
# Purpose count amongst the active students
df[df['STATUS']=='ACTIVE']['PURPOSE'].value_counts()
```

```python
df.columns
```

```python
# status count
df['STATUS'].value_counts()
```

```python
# rename the columns 'Unnamed: __'; follow uniform convention
df.rename(columns={'Unnamed: 12': 'MAY_DATE',
                   'Unnamed: 13': 'MAY_MODE',
                   'Unnamed: 15': 'JUNE_DATE',
                   'Unnamed: 16': 'JUNE_MODE',
                   'Unnamed: 18': 'JULY_DATE',
                   'Unnamed: 19': 'JULY_MODE',
                   'Unnamed: 21': 'AUG_DATE',
                   'Unnamed: 22': 'AUG_MODE',
                   'Unnamed: 24': 'SEP_DATE',
                   'Unnamed: 25': 'SEP_MODE',
                  }, inplace=True)
```

```python
df.columns
```

```python
# drop the columns >= October
df.drop(columns=['OCTOBER',
       'Unnamed: 27', 'Unnamed: 28', 'NOVEMBER', 'Unnamed: 30', 'Unnamed: 31',
       'DECEMBER', 'Unnamed: 33', 'Unnamed: 34'], inplace=True)
```

```python
df.info()
```

```python
# JULY_DATE is not in proper format
df['JULY_DATE'].unique()
```

```python
# replace the string 'MOVED TO GRC 1.0' with NaN
df.replace('MOVED TO GRC 1.0', np.nan, inplace=True) 
```

```python
df['JULY_DATE'].info()
```

```python
df.head()
```

```python
n_stud = []

for col in df.columns:
    if 'DATE' in col:
        stud_count = len(df)-df[col].isnull().sum()
        n_stud.append((col, stud_count))

df_nstud = pd.DataFrame(n_stud)
```

```python
df_nstud
```

```python
df_nstud = df_nstud.rename(columns={0: 'Month', 1: 'Strength'})
df_nstud.drop([0,1,6], inplace=True)
df_nstud['Month'] = df_nstud['Month'].replace({'MAY_DATE': 'May', 'JUNE_DATE':'June', 
                                               'JULY_DATE':'July', 'AUG_DATE':'August'})
df_nstud
```

```python
df['LEAVING MONTH'] = df[~df['LEAVING DATE'].isnull()]['LEAVING DATE'].dt.strftime('%B')
```

```python
df.head()
```

```python
df['LEAVING MONTH'].value_counts()
```

```python
month_counts = df['LEAVING MONTH'].value_counts().reset_index()
month_counts.columns = ['Month', 'Left']
```

```python
df_nstud = df_nstud.merge(month_counts, on='Month', how='left')
```

```python
df2 = df_nstud.copy()
```

```python
df2
```

```python
df2['Left'] = df2['Left'].fillna(0).astype(int)
df2['Net gain'] = df2['Strength'] - df2['Left']
df2
```

```python
df2['New'] = df2['Strength'].diff().fillna(0)
df2['Net gain'] = df2['Net gain'].diff().fillna(0)
```

```python
df2['New'] = df2['Strength'].diff().fillna(0)
df2
```
