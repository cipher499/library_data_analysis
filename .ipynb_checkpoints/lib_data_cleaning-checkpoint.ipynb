```python
df = pd.read_excel('gramsci2_c1.xlsx')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.replace('%', '',inplace=True)
df.replace('Bank ', 'Bank', inplace=True)
```

```python
df.columns
```

```python
df[df['STATUS']=='ACTIVE']['PURPOSE'].value_counts()
```

```python
df['STATUS'].value_counts()
```

```python
df.columns
```

```python
df.rename(columns={'Unnamed: 12': 'MAY_DATE',
                   'Unnamed: 13': 'MAY_MODE',
                   'Unnamed: 15': 'JUNE_DATE',
                   'Unnamed: 16': 'JUNE_MODE',
                   'Unnamed: 18': 'JULY_DATE',
                   'Unnamed: 19': 'JULY_MODE',
                   'Unnamed: 21': 'AUG_DATE',
                   'Unnamed: 22': 'AUG_MODE',
                   'Unnamed: 24': 'SEP_DATE',
                   'Unnamed: 25': 'SEP_MODE',
                  }, inplace=True)
```

```python
df.columns
```

```python
df.drop(columns=['OCTOBER',
       'Unnamed: 27', 'Unnamed: 28', 'NOVEMBER', 'Unnamed: 30', 'Unnamed: 31',
       'DECEMBER', 'Unnamed: 33', 'Unnamed: 34'], inplace=True)
```

```python
df.info()
```

```python
df[df['JULY_DATE']=='MOVED TO GRC 1.0']
df.replace('MOVED TO GRC 1.0', np.nan, inplace=True) 
```

```python
n_stud = []
for col in df.columns:
    if 'DATE' in col and 'LEAVING' not in col and 'SEP' not in col:
        stud_count = len(df)-df[col].isnull().sum()
        n_stud.append((col, stud_count))
df_nstud = pd.DataFrame(n_stud)
```

```python
df_nstud
```

```python
monthly_rev = []
for col in df[['MAY', 'JUNE', 'JULY', 'AUGUST']].columns:
    rev = df[col].sum()
    monthly_rev.append((col,rev))
```

```python
monthly_rev
```

```python
df_rev = pd.DataFrame(monthly_rev)
df_rev
```
